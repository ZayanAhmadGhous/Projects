# TITANIC DATASET FULL EDA 

```python
# First of all we have to import all the essential libraires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Now We have to load the data from seaborn and make a dataframe
df = sns.load_dataset('titanic')
```

## Expolatory Data Analysis

```python
# First of all we have to see the head of the data
df.head()
```

```python
# Now I move towards the describe function
df.describe()
```

```python
# Now I have to move towards data types of the data
df.info()
```

```python
# Now I go to shape of the data
df.shape
```

```python
# Checking is there is any null values in the data
df.isnull().sum()
```

```python
# We have too many missing values in the deck column so we remove it
df.drop('deck' , axis=1 , inplace=True)
```

```python
# There is also some missing values in the age columns so we compute it with some math operation
df['age'] = df['age'].fillna(df['age'].median())
```

```python
# There is some also missing values in embark_town it is categorical so we compute it by its mode
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])
```

```python
# Viewing the head of the data so we can have a look on the data
df.head()
```

```python
# Now we will move towards drop duplicate columns or those which are not essential for out model training
df.drop(columns=['embarked', 'adult_male', 'who', 'alive','alone','sibsp','parch','class','embark_town'] , inplace=True)
```

```python
# Checking the head of the data 
df.head()
```

```python
# checking for missing values
df.isnull().sum()
```

```python
# Mapping columns to convert into numeraical data
df['sex'] = df['sex'].map({'male':0 , 'female':1})
```
